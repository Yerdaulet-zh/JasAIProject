```python
import os, json, torch, cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from PIL import Image
from patchify import patchify, unpatchify
```

```python
val = pd.read_csv("dataset/val_annotation.csv")
test = pd.read_csv("dataset/test_annotation.csv")
train = pd.read_csv("dataset/train_annotation.csv")
train.drop(8, inplace=True)
train.drop(64, inplace=True)
train.drop(117, inplace=True)
train.drop(213, inplace=True)
```

# Patchify 

```python
# x = 660
x = 0
image = np.array(Image.open("dataset/JPEGImages/" + train.iloc[x, 1][:-3] + "jpg").convert("RGB"))
resized_image = cv2.resize(image, (112, 48))
plt.imshow(resized_image)
plt.show()
```

```python
patches = patchify(image=resized_image[:, :, 1], patch_size=(16, 16), step=16).reshape(21, -1)
patches.shape
```

```python
for patch in patches:
    plt.imshow(patch.reshape(16, 16), "gray")
    plt.show()
```
